```python
# Load dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
df=pd.read_csv('winequality_edited.csv')
df.head()
```

### Histograms for Various Features

```python
df.shape[1]
```

```python
plt.hist(df['fixed_acidity'])
plt.show()
```

```python
def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure()
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name+" Distribution")
    fig.set_figheight(15)
    fig.set_figwidth(15)    
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

#test = pd.DataFrame(np.random.randn(30, 9), columns=map(str, range(9)))
draw_histograms(df, df.columns, 5, 3)
```

### Scatterplots of Quality Against Various Features

```python
def draw_scatterplot(df, variables, n_rows, n_cols):
    fig=plt.figure()
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        #df[var_name].hist(bins=10,ax=ax)
        plt.scatter(df['quality'], df[var_name],alpha=0.5)
        ax.set_title(var_name+" Distribution")
    fig.set_figheight(15)
    fig.set_figwidth(15)    
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

#test = pd.DataFrame(np.random.randn(30, 9), columns=map(str, range(9)))
draw_scatterplot(df, df.columns, 5, 3)
```
